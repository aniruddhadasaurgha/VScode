```python
import sqlite3
import pandas as pd

db_path = "basketball.sqlite"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tables in database:")
print(tables)

team = pd.read_sql_query("SELECT * FROM Team;", conn)
print("\nTeam table:")
print(team.head())
print("\nTeam columns:")
print(team.columns)

team_attr = pd.read_sql_query("SELECT * FROM Team_Attributes;", conn)
print("\nTeam_Attributes table:")
print(team_attr.head())
print("\nTeam_Attributes columns:")
print(team_attr.columns)

query = """
SELECT *
FROM Team_Attributes
WHERE ID IN (
    SELECT id
    FROM Team
    WHERE state = 'New York'
);
"""
ny_team_attr = pd.read_sql_query(query, conn)

print("\nTeam_Attributes for teams in New York:")
print(ny_team_attr)

conn.close()
```
