# **Opening the file using with() function and giving my introduction**

```python
with open("Codingal.txt", "w") as f:
  f.write("Hi! My name is Aniruddha Das Aurgha")
  f.close()
```

#**Splitting the words**

```python
with open("Codingal.txt", "r") as f:
  data = f.readlines()
  for line in data:
    word = line.split()
    print(word)
    f.close()
```

# **Checking if My_File.txt exists or not**

```python
import os
if os.path.exists("My_File.txt"):
  print("File exists")
else:
  print("File does not exist")
```
