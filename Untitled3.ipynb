# **Printing the contents of the file**

```python
file = open("Sample_File.txt", "r")
print(file.read())
file.close()

```

# **Printing first 10 charecters**

```python
file = open("Sample_File.txt", "r")
print(file.read(10))
file.close()

```

# **Printing the first line**

```python
file = open("Sample_File.txt", "r")
print(file.readline())
file.close()

```

# **Printing the first 4 lines**

```python
file = open("Sample_File.txt", "r")
print(file.readline())
print(file.readline())
print(file.readline())
print(file.readline())
file.close()

```

# **Looping through the contents and printing all the contents of the file one by one**

```python
file = open("Sample_File.txt", "r")
for x in file:
  print(x)
file.close()

```
